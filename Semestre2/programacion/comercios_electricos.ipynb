```python
import pandas as pd
import geopandas as gpd
import folium
import matplotlib.pyplot as plt
```

```python
# los comercios se descargan de https://www.inegi.org.mx/app/descarga/?ti=6
df = pd.read_csv('denue_inegi_43_.csv', encoding='ISO-8859-1')
```

```python
df.head()
```

```python
print(df['nombre_act'].unique())
```

```python
df_mat_elec = df[df['nombre_act'] == 'Comercio al por mayor de equipo y material eléctrico']
```

```python
df_mat_elec.head()
```

```python
# group the data by the 'Country' column and calculate the mean age
grouped = df_mat_elec.groupby('entidad')['id'].count()

# create a new dataframe based on the grouped data
new_df = pd.DataFrame({
    'entidad': grouped.index,
    'Comercios': grouped.values
})

print(new_df)
```

```python
df_mat_elec.describe()
```

```python
# create a map centered on Mexico
mexico_map = folium.Map(location=[23.6345, -102.5528], zoom_start=5)

# add markers for each store location
for index, row in df_mat_elec.iterrows():
    folium.Marker(
        location=[row['latitud'], row['longitud']],
        popup=row['id'],
        icon=folium.Icon(color='red', icon='shopping-cart')
    ).add_to(mexico_map)

# display the map
mexico_map
```
