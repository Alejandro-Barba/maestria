#### This data set contains 50 samples of each of three species of Iris:
Iris setosa, Iris virginica and Iris versicolor, for each of these species four traits were
measured for each sample: the length and width of the sepals and petals. 

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# fits i read the csv and save it to the df variable
df = pd.read_csv('Iris.csv')
df.head()
```

```python
# Describing the fields of the dataset
df.describe()
```

```python
# Describing the Iris setosa
df[(df.Species == "Iris-setosa")].describe()
```

```python
# Describing the Iris versicolor
df[(df.Species == "Iris-versicolor")].describe()
```

```python
# Describing the Iris virginica
df[(df.Species == "Iris-virginica")].describe()
```

## Plot the iris dataset: Boxplot, Histogram Plot, Violin Plot, Scatter Plot, Andrews curve and proposed one multivariate graphical representation.

```python
# I¬¥m using the boxplot to check the distribution of the sepal lenght
sns.boxplot(x="Species", y="PetalLengthCm", data=df)
```

```python
# I¬¥m using the boxplot to check the distribution of the sepal Width
plt.hist(df['SepalWidthCm'])
```

```python
# I¬¥m using the violin plot to check the distribution of the sepal Width

sns.violinplot(x="Species", y="PetalLengthCm", data=df)
```

## As we can see the petal length for the setosa is really diferent from the other 2 species, we can identify them very easily, for the Versicolor and the virginica some values are overlapping making them more dificult to identify

```python
#Scatter plot to showthe distribution of the petal width
plt.scatter(df['PetalWidthCm'],df['Id'])
```

```python
#Andrews curve
pd.plotting.andrews_curves(df.drop("Id", axis=1), "Species")
```

```python
#Multivariable graph representations
ax = df[['SepalLengthCm', 'SepalWidthCm','PetalLengthCm','PetalWidthCm']].plot(kind='box', title='boxplot')
df.drop("Id", axis=1).boxplot(by="Species")
plt.show()
```

## 2.- Respond ‚ÄúTrue or False‚Äù and provide a numerical example that provide support to the response:
- (a) Mean is robust against outliers.
- (b) Median is robust against outliers.
- (c) Standard deviation is robust against outliers.

```python
# A is false, the mean is easily changed by altering only 1 single outlier

data = (69,74,68, 70, 72, 67,66,70,76,68, 72, 79, 74,67,66,71,74, 75, 75, 76)
data2 = (69,74,68, 70, 72, 67,66,70,76,68, 72, 79, 74,67,66,71,74, 75, 75, 500)

print(f'The original mean is {np.mean(data)}dd')
print(f'The modified mean is {np.mean(data2)}dd')
```

```python
# B is true, the median is robust against outliers

data = (69,74,68, 70, 72, 67,66,70,76,68, 72, 79, 74,67,66,71,74, 75, 75, 76)
data2 = (69,74,68, 70, 72, 67,66,70,76,68, 72, 79, 74,67,66,71,74, 75, 75, 500)

print(f'The original median is {np.median(data)}dd')
print(f'The modified median is {np.median(data2)}dd')
```

```python
# C is false, the strandar deviation is easily changed by altering only 1 single outlier

data = (69,74,68, 70, 72, 67,66,70,76,68, 72, 79, 74,67,66,71,74, 75, 75, 76)
data2 = (69,74,68, 70, 72, 67,66,70,76,68, 72, 79, 74,67,66,71,74, 75, 75, 500)

print(f'The original sdt is {np.std(data)}dd')
print(f'The modified sdtd is {np.std(data2)}dd')
```
